```
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow import keras
from google.colab import files
from io import BytesIO
from PIL import Image
from tensorflow.python.keras.preprocessing import image as kp_image
```

```
model = keras.applications.VGG16() # > 500 MB
```

```
uploaded = files.upload()
```

```
img = Image.open(BytesIO(uploaded['car-for-whole-life.jpg']))
plt.imshow(img)
```

```
img = np.array(img) # преобразование в массив numpy
x = keras.applications.vgg16.preprocess_input(img) # получить нужный массив в нужном цветовом формате
print(x.shape)
x = np.expand_dims(x, axis=0) # добавим первую пространственную ось
print(x.shape)
```

```
res = model.predict(x) # (batch_size, height, width, channels)
print(np.argmax(res))
```
